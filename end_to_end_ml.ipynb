```python
import pandas as pd
import numpy as np
df=pd.read_csv("https://github.com/campusx-official/placement-project-logistic-regression/raw/refs/heads/main/placement.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.iloc[:,1:] # removing Top upper unwanted row
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
```

```python
# Apply logical Regrarion
```

```python
x=df.iloc[:,0:2] # Sara row but column 0 -2 matlab 0 and 1
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
y.shape # x is 2d tensor and y is 1d tensor
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1) # x is independent and y is dependent variable
# 0.1 data train test ma chalajayanga (90) x and y sa
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test # y ka sample data sa pata chalaga placement hoga ya nahi
```

```python
from sklearn.preprocessing import StandardScaler # for scale the value b/w -1 to 1
# sklearn library ma preprocessing ma jayanga usma class hai StandardScaler is class ka obj necha
```

```python
scaler=StandardScaler()
```

```python
# scale the data all value b/w -1 to 1
x_train = scaler.fit_transform(x_train) # fit_transform samjhaga ki x data ma kia hai fir usa transform kardega
```

```python
x_train
```

```python
x_test = scaler.transform(x_test) # upper ma fit use kar liya than no need dubara
```

```python
x_test
```

```python
# Model Train using logistic regration classifire
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train) # here fit use for train model
```

```python
# Model evaluation ma package nikal lata hai test data per
y_pred=clf.predict(x_test) # my model say
```

```python
y_test # actual data
```

```python
from sklearn.metrics import accuracy_score # 0.7 means 10 input dena per 7 correct predict kar ka de raha hai
accuracy_score(y_test,y_pred) # accuracy_score ma sahi data and predict wala dalta hai
```

```python
from ast import pattern
# For chaking ml model na kia pattern notice kia data ma usko visullize karna ka liya use decesion boundary iska lya library import mlxtend (plot)
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train,y_train.values,clf=clf, legend=2) # .values use for contverting numpy array x phale hogya tha Standerdscaler ma dalna per and clf= clf b/c classifier ka name hai classifire
```
